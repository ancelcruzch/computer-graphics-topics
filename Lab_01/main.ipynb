```python
#Generar el histograma de la imagen original
image_hist_original, bins = np.histogram(image.flatten(), bins=256, range=[0,256])
plt.title('Histograma Original ')
plt.plot(image_hist_original, color='black')
```

```python
# Mostrar los niveles de gris presentes en la imagen
levels = np.where(image_hist_original > 0)[0]  # Los niveles de gris con frecuencia mayor que 0
frequencies = image_hist_original[levels]  # Las frecuencias correspondientes

# Total de pixeles
height, width = image.shape[:2]
total_pixeles = height * width
print(f"Total de pxeles en la imagen: {total_pixeles}")

# Imprimir los niveles de gris y sus frecuencias
print("Niveles de gris presentes en la imagen y sus frecuencias:")

# Contar niveles
total_niveles = 0
for level, freq in zip(levels, frequencies):
    total_niveles +=1
    print(f"Nivel de gris: {level}, Frecuencia: {freq}, Prob.:{freq/total_pixeles}")

print(f"Total de niveles de gris: {total_niveles}")
```

```python
print(f"Total de niveles de gris: {total_niveles}")
# Mejorar niveles de gris
prob = 0 
for level, freq in zip(levels, frequencies):
    prob += freq / total_pixeles
    print(f"Nivel de gris: {level} , S_k.: {total_niveles * prob}")
```

```python
# Inicializamos la lista para guardar los valores de S_k
sk_values = []

# Calcular los valores de S_k y guardarlos
prob = 0
for level, freq in zip(levels, frequencies):
    prob += freq / total_pixeles
    sk_value = total_niveles * prob
    sk_values.append(sk_value)
    print(f"Nivel de gris: {level} , S_k.: {sk_value}")

# Generar el histograma de los valores de S_k
plt.figure(figsize=(10, 5))
plt.hist(sk_values, bins=256, color='black', alpha=0.75)
plt.title('Histograma de los Valores de S_k')
plt.xlabel('Valores de S_k')
plt.ylabel('Frecuencia')
plt.show()
```
